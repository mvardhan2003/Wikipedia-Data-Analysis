```python
cols = ['Articles', 'All pages','Edits', 'Users', 'Active users', 'Files', 'Depth']
cols
```

```python
for c in cols:
    df[c] = df[c].astype(str).str.replace(',','',regex=False).astype(int)
df.info()
```

```python
df.columns
```

```python
for c in ["Language","Language (local)","Wiki"]:
    df[c] = df[c].astype(str)
df.info()
```
