```python
ART_TH = df['Articles'].median()
DEP_TH = df['Depth'].median()       
```

```python
def label_row(row, art_th=ART_TH, dep_th=DEP_TH):
    high_art = row['Articles'] >= art_th
    high_dep = row['Depth'] >= dep_th
    if high_art and high_dep:
        return 'High_Qty_High_Quality'
    if high_art and not high_dep:
        return 'High_Qty_Low_Quality'
    if not high_art and high_dep:
        return 'Low_Qty_High_Quality'
    return 'Low_Qty_Low_Quality'
```

```python
df['quality_quantity'] = df.apply(label_row, axis=1)
df
```

```python
df.to_csv("Wikipedia_processed.csv")
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x='Articles', y='Depth',data=df,hue='quality_quantity',            
    palette={
        'High_Qty_High_Quality': 'green',
        'High_Qty_Low_Quality': 'blue',
        'Low_Qty_High_Quality': 'purple',
        'Low_Qty_Low_Quality': 'grey'
    }
)
plt.xscale('log')
plt.yscale('log')
plt.savefig("results/Quality vs Quantity Classification of Wikipedia Editions (Log–Log Scale).png")
```

From the analysis, we observed that Wikipedia language editions differ not only in how many articles they contain (quantity) but also in how well those articles are developed (quality). By using Articles and Depth metrics, we divided the wikis into four groups and clearly saw that some languages have very high article counts but low depth (e.g., Cebuano, Swedish), while others balance both high quantity and high quality (e.g., English, French, Spanish).

This proves that article count alone cannot represent the true strength of a Wikipedia edition. Depth is equally important, and the Quality vs Quantity classification helps us understand how each language edition grows and where improvement is needed.
